```python
fortune = pd.read_csv(r'C:\Users\pc\Downloads\fortune1000.csv', index_col=['Rank'])

fortune['Sector'] = fortune['Sector'].astype('category')
sector = fortune.groupby('Sector')

df = pd.DataFrame(columns=fortune.columns)

for sec, data in sector:
    hp = data.nlargest(1, "Profits")
    df = pd.concat([df, hp], ignore_index=True)  # Fix here

sns.barplot(x=df['Sector'], y=df['Profits'], data=df)
plt.xticks(rotation=90)
```
