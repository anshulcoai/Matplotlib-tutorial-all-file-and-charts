```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tip=sns.load_dataset('tips')
sns.swarmplot(x='day',y='total_bill',data=tip)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tip,size=5)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tip,orient='v')
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tip)
```

```python
np.random.seed(1)
l=np.random.randn(1000)
rs=pd.Series(l,name='Random Element')
sns.histplot(data=rs,kde=True)
```
