```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=np.array([1,2,3,4,])
Y=X*2
plt.plot(X,Y)
```

```python
X=np.array([1,2,3,4])
Y=X*2
plt.plot(X,Y)
plt.xlabel("X data")
plt.ylabel("Result X*2")
plt.title('Random Data to show Relation between X and X*2')
```

```python
X=np.array([1,2,3,4,])
Y=X*2
plt.figure(figsize=(6,6))
plt.plot(X,Y, marker='o',linestyle='-')
for i, (X1,Y1) in enumerate (zip(X,Y)):
    plt.annotate(f'({X1},{Y1})',(X1,Y1),textcoords='offset points',
                xytext=(0,10),ha='center')
plt.xlabel("X data")    
plt.ylabel("Y data")
plt.title("Random Data to show relation between X and X*2")
```

```python
X=np.array([1,2,3,4,])
Y=X*2
plt.plot(X,Y)
plt.xlabel("X data")
plt.ylabel("Result X*2")
plt.title('Random Data to show Relation between X and X*2')

Z=X*3
plt.figure(figsize=(6,6))
plt.plot(X,Z, marker='o',linestyle='-')
for i, (X1,Z1) in enumerate(zip(X,Z)):
    plt.annotate(f'({X1},{Z1})',(X1,Z1),textcoords='offset points',
                 xytext=(0,10),ha='center')
plt.xlabel("X data")    
plt.ylabel("Result X*3")
plt.title('Random Data to show Relation between X and X*3')
```

```python
X=np.array([1,2,3,4,])
Y=X*2
plt.plot(X,Y)

Z=X*3
plt.plot(X,Z,marker='o',linestyle='-')
for i, (X1,Z1) in enumerate(zip(X,Z)):
    plt.annotate(f'({X1},{Z1})',(X1,Z1),textcoords='offset points',
                xytext=(0,10),ha='center')

plt.fill_between(X,Y,Z,color='black')
plt.xlabel("X data")
plt.ylabel("REsult of X*3")
plt.title('Random Data to show Relation between X and X*3')
```

```python
fortune = pd.read_csv('C:/Users/pc/Downloads/fortune1000.csv', index_col='Rank')
fortune['Sector']=fortune['Sector'].astype('category')
sector=fortune.groupby('Sector')
df=sector.agg({'Revenue':'sum','Profits':'sum'})
df=df.head(6)
plt.figure(figsize=(10,10))
plt.bar(df.index,df['Profits'])
plt.xlabel("Sector Name")
plt.ylabel("Profits")
```
