```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars={'audi':25,'BMW':41,'ford':26,'Tesla':48,'Jaguar':12,'Mercedes':36,'Toyota':25}
explode=(0.0,0.0,0.0,0.3,0.0,0.0,0.0)
colors=('grey','magenta','brown','red','yellow','pink','indigo')
wp={'linewidth':3,'edgecolor':'black'}

def func(pct, data):
    val = int(pct / 100 * np.sum(list(data)))
    return "{:.2f}%\n({:d})".format(pct, val)
fig, ax = plt.subplots(figsize=(10, 8))
wedges, texts, autotexts = ax.pie(cars.values(),
                                 autopct=lambda pct: func(pct, cars.values()),
                                 explode=explode, labels=cars.keys(),
                                 colors=colors,startangle=90, wedgeprops=wp,
                                 textprops=dict(color='blue'))
ax.legend(wedges, cars.keys(), title='Cars Sale report', loc='center left',
         bbox_to_anchor=(1,0,0.7,1))
ax.set_title('Sale Report for Luxury Cars')
```
