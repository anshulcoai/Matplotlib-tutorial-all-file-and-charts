```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\pc\Downloads\salesman.csv',parse_dates=['Date'])
```

```python
sales_data = df.groupby("Salesman")["Revenue"].sum()

plt.pie(sales_data, labels=sales_data.index, autopct='%1.1f%%')
plt.title("Sales Revenue Distribution by Salesman")
plt.show()
```

```python
food=pd.read_csv(r'C:\Users\pc\Downloads\foods.csv')
```

```python
food
```

```python
food['Item'].value_counts()
```

```python
avg_spent=food.groupby('Item')["Spend"].mean()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(avg_spent.index, avg_spent, color="teal")
plt.xlabel("Food Items")
plt.ylabel("Average Amount Spent")
plt.title("Average Spending on Each Food Item")

plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(avg_spent, labels=avg_spent.index, autopct='%1.1f%%', startangle=140)
plt.title("Item-wise Average Spending")
plt.legend()
plt.show()
```
