```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mw=np.array([160,165,170,175,180,185,190,195,200,205])
mh=np.array([55,58,60,62,64,66,68,70,72,74])

fw=np.array([150,155,160,165,170,175,180,195,200,205])
fh=np.array([50,52,54,56,58,64,66,68,70,72])

plt.scatter(mw,mh,color='blue',label='Male')
plt.scatter(fw,fh,color='red',label='Female')

plt.xlabel('Height in cm')
plt.ylabel('Weight in Kg')

plt.title('Comparison of Height Weight')
plt.legend()
plt.show()
```
