```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=np.random.randn(1000)
plt.hist(data,bins=50,color='r',edgecolor='b')
plt.xlabel('Random elements')
plt.ylabel('Frequncy Count')
plt.title('Frequency Count')
```

```python
pokemon=pd.read_csv(r'C:\Users\pc\Downloads\pokemon123.csv',
                   index_col='name').squeeze()
```

```python
pokemon.head(50)
```

```python
df=pokemon.value_counts()

plt.figure(figsize=(10,10))
plt.bar(df.index, df.values, color='green', edgecolor='yellow')

plt.xlabel('pokemon type')
plt.ylabel('pokemon Frequency Count')
plt.xticks(rotation=45)

plt.show()
```
